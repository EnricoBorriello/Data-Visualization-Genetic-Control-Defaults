# Figure 1

```python
data = pd.read_csv('data_files/figure_1.csv', 
                   names=['size', 'bacteria', 'archaea','eukarya','virus'])[1:].astype(float)
```

```python
data[:4]
```

```python
from scipy.interpolate import interp1d
```

```python
size_log = np.log10(np.array(data['size']))
bac_num = np.array(data['bacteria'])
arc_num = np.array(data['archaea'])
euk_num = np.array(data['eukarya'])
vir_num = np.array(data['virus'])
```

```python
# interpolation in the linear-log scales

bac_interp_lin_log = interp1d(size_log,bac_num,kind='cubic')
arc_interp_lin_log = interp1d(size_log,arc_num,kind='cubic')
euk_interp_lin_log = interp1d(size_log,euk_num,kind='cubic')
vir_interp_lin_log = interp1d(size_log,vir_num,kind='cubic')
```

```python
fig, ax = plt.subplots()
plt.xscale('log')

x = 10**np.arange(-3,5,.01)

ybac = bac_interp_lin_log(np.log10(x))
yarc = arc_interp_lin_log(np.log10(x))
yeuk = euk_interp_lin_log(np.log10(x))
yvir = vir_interp_lin_log(np.log10(x))

plt.plot(x,yvir,label=None,linewidth=6,c='white') # decorative
plt.plot(x,yvir,label='Virus',linewidth=2,color='tab:green')

plt.plot(x,yarc,label=None,linewidth=6,c='white') # decorative
plt.plot(x,yarc,label='Archaea',linewidth=2,c='darkgray')

plt.plot(x,ybac,label=None,linewidth=6,color='white') # decorative
plt.plot(x,ybac,label='Bacteria',linewidth=2,c='orange')

plt.plot(x,yeuk,label=None,linewidth=6,color='white') # decorative
plt.plot(x,yeuk,label='Eukarya',linewidth=2,color='tab:blue')


plt.plot([10**(-3),10**4],[0,0],c='white',linewidth=7.6)

plt.ylim(0.00,.5);
plt.xlim(10**(-3),10**4)

text_size = 14

plt.xlabel('genome size (megabases)',size = 1.0*text_size)
plt.ylabel('relative number of genomes',size = 1.0*text_size)

plt.tick_params(axis=u'both', labelsize=text_size)

# GRIDLINES
#ax.minorticks_on()
ax.yaxis.grid(True, linestyle='-',c='whitesmoke')
ax.xaxis.grid(True, linestyle='-',c='whitesmoke',)
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2);

# LEGEND
plt.legend(prop={"size":.8*text_size},fancybox=True, framealpha=1, shadow=True, borderpad=.5);

# EXPORT
plt.savefig("figure_1.pdf", bbox_inches = 'tight',pad_inches = .1,dpi=100)
```

# Figure 3D

```python
data = pd.read_csv('data_files/figure_3D.csv')
                  #names=['bug1', 'bug2', 'bug3','bug4','bug4'])[1:].astype(float)
```

```python
fit1 = np.array(data['Organism_1'])
fit2 = np.array(data['Organism_2'])
fit3 = np.array(data['Organism_3'])
fit4 = np.array(data['Organism_4'])
fit5 = np.array(data['Organism_5'])
```

```python
import matplotlib.pylab as pl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = 100.*np.arange(len(fit1))/(10**6)

y1  = np.ones(x.size)
y2  = np.ones(x.size)*2
y3  = np.ones(x.size)*3
y4  = np.ones(x.size)*4
y5  = np.ones(x.size)*5

z1 = fit1
z2 = fit2
z3 = fit3
z4 = fit4
z5 = fit5

pl.figure()
ax = pl.subplot(projection='3d')

ax.plot(x, y5, z5, color='w',linewidth=4)
ax.plot(x, y5, z5, color='black')

ax.plot(x, y4, z4, color='w',linewidth=4)
ax.plot(x, y4, z4, color='darkgray')

ax.plot(x, y3, z3, color='w',linewidth=4)
ax.plot(x, y3, z3, color='tab:green')

ax.plot(x, y2, z2, color='w',linewidth=4)
ax.plot(x, y2, z2, color='orange')

ax.plot(x, y1, z1, color='w',linewidth=4)
ax.plot(x, y1, z1, color='tab:blue')




ax.view_init(30, -75);

text_size = 18

ax.set_xlabel('time (millions of generations)',size=text_size, labelpad=16);
ax.set_ylabel('organism',size=text_size, labelpad=16);
ax.set_zlabel('(individual) fitness',size=text_size, labelpad=16);




# Get rid of colored axes planes
# First remove fill
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

# Now set color to white (or whatever is "invisible")
ax.xaxis.pane.set_edgecolor('w')
ax.yaxis.pane.set_edgecolor('w')
ax.zaxis.pane.set_edgecolor('w')

ax.set_yticks(range(1,6,1));

pl.tight_layout()

pl.tick_params(axis=u'x', labelsize=text_size)
pl.tick_params(axis=u'y', labelsize=text_size)
pl.tick_params(axis=u'z', labelsize=text_size)

#pl.tight_layout(pad=.10, w_pad=10, h_pad=10)
pl.subplots_adjust(left=4, bottom=4, right=5.5, top=5.5)

ax.set_zlim(0,30)
ax.set_xlim(0,.85)

# EXPORT
plt.savefig("figure_3d.pdf", bbox_inches = 'tight',pad_inches = .1,dpi=100)
```
